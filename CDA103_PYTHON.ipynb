The reason we use a temp variable is to prevent data loss. Let's say you have two variables, x and y, with initial values.

If you tried to swap them directly like this:

x = y (The original value of x would be overwritten and lost)
y = x (Now y would just get the new value of x, which is the original value of y, so both x and y would end up with the same value as y's original value)
To avoid this, the temp variable acts as a temporary storage:

temp = x: We first store the original value of x in temp.
x = y: Now, we can safely assign the value of y to x, because x's original value is preserved in temp.
y = temp: Finally, we assign the value from temp (which was the original value of x) to y.
This three-step process ensures that both original values are maintained and correctly swapped. While Python offers a more concise way to swap variables using tuple assignment (x, y = y, x, as shown in cell -46loH1zQDYS), understanding the temp variable method is fundamental to computer science concepts of variable assignment and memory management.


````python
x = 5
y = 7
temp = x #temp = 5 ,x = 5 ,y = 7
x = y #temp = 5, x = 7, y = 7
y = temp #temp = 5, y = 5, x = 7
print("after swappin:", x,y)
````
